# Link to Bills web scrape 

```python
 soup = BeautifulSoup(source.text, 'html.parser')

    # Find all tables with a class containing 'wikitable'
tables = soup.find_all('table', class_=lambda x: x and 'wikitable' in x)

    # Check if any tables are found
if tables:
        # Create an empty list to store DataFrames for each table
        dfs = []

        for table in tables:
            # Access the tbody of the table
            tbody = table.tbody

            # Create a list to store rows of data for the current table
            table_data = []

            # Now you can iterate through the rows in the tbody
            for row in tbody.find_all('tr'):
                # Extract data from each cell in the row
                cells = row.find_all(['td', 'th'])
                if cells:
                    # Extract text content and links from cells
                    data = [cell.text.strip() if not cell.find('a') else cell.find('a')['href'] for cell in cells]
                    table_data.append(data)

            # Convert the list of rows into a pandas DataFrame
            df = pd.DataFrame(table_data[1:], columns=table_data[0])

            # Append the DataFrame to the list
            dfs.append(df)

        # Concatenate all DataFrames in the list
        final_df = pd.concat(dfs, ignore_index=True)

        # Display the final DataFrame
        print(final_df)
```
